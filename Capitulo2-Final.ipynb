```python
from sklearn.datasets import load_diabetes
import matplotlib.pyplot as plt
```

```python
# Cargar información
diabetes = load_diabetes()

X = diabetes.data
y = diabetes.target
```

```python
#Ver información
AGE = 0; SEX= 1; MASS_INDEX = 2; BLOOD_PRESURE = 3;
print(diabetes.DESCR)
```

```python
print(X.shape) #-> 442 columnas y 10 filas
print(y.shape) # -> Adecuado para regresión.
print(X[0:5,:])
print(y[0:5])
print(X[1,MASS_INDEX])
print(X[1:3,AGE])
print(diabetes.data.shape[0])
```

```python
x = range(0, diabetes.data.shape[0],1)  #- > lo mismo con x= range(0, len(y), 1)
```

```python
plt.figure(figsize=(30,8))
plt.plot(x, X[:, MASS_INDEX], 'y-', label='MASS Index')
plt.plot(x, X[:, BLOOD_PRESURE], 'b-', label='Blood Presure')
plt.rcParams.update({'font.size':22})
plt.legend(prop={'size':25}); plt.grid(); plt.show()
```

```python
fig, axs = plt.subplots(1,2, figsize=(12,4))
for fig, feature, label in zip(range(2), [MASS_INDEX,AGE], ['Mass index','Age']):
    axs[fig].scatter(X[:,feature], X[:, BLOOD_PRESURE], c='g', marker='o')
    axs[fig].set_xlabel(label); axs[fig].set_ylabel('Blood Pressure')
    axs[fig].grid();
```
